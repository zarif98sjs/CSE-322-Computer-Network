# (1) Node

```python
file_names = ['lrwpan-npan-3-nflow-10-app-3000bps',
'lrwpan-npan-6-nflow-10-app-3000bps',
'lrwpan-npan-10-nflow-10-app-3000bps',
'lrwpan-npan-13-nflow-10-app-3000bps',
'lrwpan-npan-18-nflow-10-app-3000bps']

labels = ['node count : 8 (nLeaf = 3*2)',
'node count : 14 (nLeaf = 6*2)',
'node count : 22 (nLeaf = 10*2)',
'node count : 28 (nLeaf = 13*2)',
'node count : 38 (nLeaf = 18*2)']

for i  in range(len(file_names)):
    file_name = file_names[i]
    label_ = labels[i]
    with open(file_name+'/throughput.dat') as file:
        lines = file.readlines()
    X = []
    Y = []
    for line in lines:
        words = line.split()
        words[0] = words[0].replace("ns","")
        words[0] = float(words[0])/(1e9)
        words[1] = float(words[1])
        X.append(words[0])
        Y.append(words[1])
        # print(words)
    # print(X)
    # print(Y)
    plt.rcParams["figure.figsize"] = (12,8)
    plt.xlabel('time (second)')
    plt.ylabel('throughput (bit/s)')
    plt.title('Network Throughput')
    plt.plot(X,Y,label=label_)
    plt.legend()
```
