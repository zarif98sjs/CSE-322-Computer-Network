```python
file_names = ['wifi-nwifi-3-nflow-3-app-10Mbps-speed-10',
'wifi-nwifi-6-nflow-6-app-10Mbps-speed-10',
'wifi-nwifi-10-nflow-10-app-10Mbps-speed-10',
'wifi-nwifi-13-nflow-13-app-10Mbps-speed-10',
'wifi-nwifi-18-nflow-18-app-10Mbps-speed-10']

labels = ['node count : 8 (nLeaf = 3*2)',
'node count : 14 (nLeaf = 6*2)',
'node count : 22 (nLeaf = 10*2)',
'node count : 28 (nLeaf = 13*2)',
'node count : 38 (nLeaf = 18*2)']

for i  in range(len(file_names)):
    file_name = file_names[i]
    label_ = labels[i]
    with open(file_name+'/throughput.dat') as file:
        lines = file.readlines()
    X = []
    Y = []
    for line in lines:
        words = line.split()
        words[0] = words[0].replace("ns","")
        words[0] = float(words[0])/(1e9)
        words[1] = float(words[1])/(1e6)
        X.append(words[0])
        Y.append(words[1])
        # print(words)
    # print(X)
    # print(Y)
    plt.rcParams["figure.figsize"] = (12,8)
    plt.xlabel('time (second)')
    plt.ylabel('throughput (Mbit/s)')
    plt.title('Network Throughput')
    plt.plot(X,Y,label=label_)
    plt.legend()
```
