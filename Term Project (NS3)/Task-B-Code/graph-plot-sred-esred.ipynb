```python
import matplotlib.pyplot as plt
import numpy as np
```

# (0) Queue Drop Rate

```python
file_names = ['wired-red-nleaf-10-app-650Mbps','wired-stabilized-red-nleaf-10-app-650Mbps']
labels = ['red','sred']

for i  in range(len(file_names)):
    file_name = file_names[i]
    label_ = labels[i]
    with open(file_name+'/queueDropRate.dat') as file:
        lines = file.readlines()
    X = []
    Y = []
    for line in lines:
        words = line.split()
        words[0] = float(words[0])
        words[1] = float(words[1])
        X.append(words[0])
        Y.append(words[1])
        # print(words)
    # print(X)
    # print(Y)
    plt.rcParams["figure.figsize"] = (12,8)
    plt.xlabel('time (second)')
    plt.ylabel('queue drop rate (%)')
    plt.title('Queue Drop Rate')
    plt.fill_between(X,Y,label=label_)
    plt.legend()
```
